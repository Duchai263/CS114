```python
import yaml
import os

dataset_info = {
    'train': '/content/data/train/images',
    'val': '/content/data/val/images',
    'test': '/content/data/test/images',
    'path': os.path.abspath('/content/data'),
    'nc': 4,
    'names': ['rider', 'no_helmet', 'license_plate', 'helmet']
}

yaml_filepath = '/content/data/data.yaml'


with open(yaml_filepath, 'w') as f:
  doc = yaml.dump(
      dataset_info,
      f,
      default_flow_style=None,
      sort_keys=False
  )

```
